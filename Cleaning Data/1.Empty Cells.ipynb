```python
df = pd.read_csv('data.csv')
df.dropna(inplace = True)
print(df.to_string())
```

```python
df = pd.read_csv('data.csv')
df.fillna(130, inplace = True)
print(df.to_string())
```

```python
df = pd.read_csv('data.csv')
df["Calories"].fillna(130, inplace= True)
print(df.to_string())
```

```python
df = pd.read_csv('data.csv')
x = df['Calories'].mean()
df["Calories"].fillna(x, inplace = True)
print(df.to_string())
```
